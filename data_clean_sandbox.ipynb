```python
# libraries

import pandas
import random
import datetime

import classes
import IO
from interface import get_orbit_from, check_validity, plot_data

import numpy as np

import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.colors as colors
import matplotlib.dates

from prepare_data import get_data, display_data
```

```python
orbit = 1928 #random.randint(10,3190) 
if True:
    print('orbit :{}'.format(orbit))

datapath='/Users/alexanderwolff/Local/processed_data'


# there are times when the IMA data is not segmented into 16 periodic signals (eg. 2608)
```

```python
# choose 'ELS' or 'IMA' as your instrument
instrument = 'IMA' 

# get the data : data is stored as a pandas dataframe (read up on it if needed)
dataset = get_data(orbit, instrument, datapath, preview = True, debug=False, verbose=False)
```

```python
# example: displaying the data as an image

if dataset != (None, None):
    
    if instrument == 'MAG':
        dataset['default'].plot(figsize=(20,10))
    
    else:
        try:
            multiplier = 16 if instrument == 'IMA' else 1
            display_data(dataset['default'], dataset['sample_rate']*multiplier)
        except:
            display_data(dataset['raw'], dataset['sample_rate'])

        if instrument == 'IMA':
            print("ALL IMA STREAMS (16 orientations)")
            for stream in dataset['streams']:
                display_data(stream, dataset['sample_rate']*multiplier)

        if instrument == 'ELS':
            try:
                for stream in dataset['streams']:
                    display_data(stream, dataset['sample_rate'])
            except:
                pass

    
else:
    print('Error: Dataset does not exist.')
```

```python
# plotting various linear traces
data = dataset['default']
transposed = data.T
T = data.index

plt.figure(figsize=(20,10))
trace = transposed.max()
plt.plot(trace/trace.max())
plt.xlim([T[0], T[len(T)-1]])
plt.show()

plt.figure(figsize=(20,10))
trace = transposed.mean()
plt.plot(trace/trace.max())
plt.xlim([T[0], T[len(T)-1]])
plt.show()

plt.figure(figsize=(20,10))
trace = transposed.var()
plt.plot(trace/trace.max())
plt.xlim([T[0], T[len(T)-1]])
plt.show()
```
