```python
ntraces = list()

for i in range(16):
    ntraces.append(list())
    
    for j, value in enumerate(trace.values):
        
        if j % 16 - i == 0:
            ntraces[i].append(value)
```

```python
plt.figure(figsize=(20,10))
for i in range(16):
    
    plt.plot(ntraces[i]/max(ntraces[i]))
plt.show()
```

```python
mask = [True if value > (trace.values.max()/100.0)*10 else False for value in trace.values]

nvalues = list()
nindex = list()

for i, value in enumerate(trace.values):
    
    if mask[i]:
        nvalues.append(trace.values[i])
        nindex.append(trace.index[i])

ntrace = pandas.DataFrame(nvalues, index=nindex)
plt.figure(figsize=(20,10))
plt.plot(ntrace/ntrace.max())
plt.show()
```

```python
new_values = list()
new_times = list()

for j, value in enumerate(dataframe.to_numpy()):
    
    if j%16 == 0:
        new_values.append( value )
        new_times.append( dataframe.index[j] )


for i in range(1,16):  
    index = 0
    for j, value in enumerate(dataframe.to_numpy()):
    
        if j%16-i == 0:
            new_values[index] += value
            index+=1
            
new_values = np.stack( new_values, axis=0 )
new_dataframe = pandas.DataFrame( new_values, index = new_times )
```

```python
display_data(new_dataframe, sample_rate*16)
```
