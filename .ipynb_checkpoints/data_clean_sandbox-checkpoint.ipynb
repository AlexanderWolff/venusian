```python
# plotting various linear traces
data = dataset['default']
transposed = data.T
T = data.index

plt.figure(figsize=(20,10))
trace = transposed.max()
plt.plot(trace/trace.max())
plt.xlim([T[0], T[len(T)-1]])
plt.show()

plt.figure(figsize=(20,10))
trace = transposed.mean()
plt.plot(trace/trace.max())
plt.xlim([T[0], T[len(T)-1]])
plt.show()

plt.figure(figsize=(20,10))
trace = transposed.var()
plt.plot(trace/trace.max())
plt.xlim([T[0], T[len(T)-1]])
plt.show()
```
