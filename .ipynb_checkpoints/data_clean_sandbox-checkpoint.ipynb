```python
from sklearn import preprocessing

min_max_scaler = preprocessing.MinMaxScaler()
s = min_max_scaler.fit_transform(dataset['IMA']['default'].T)
dataset['IMA']['scaled'] = pandas.DataFrame(s).T
dataset['IMA']['scaled'].index = dataset['IMA']['default'].index

min_max_scaler = preprocessing.MinMaxScaler()
s = min_max_scaler.fit_transform(dataset['ELS']['default'].T)
dataset['ELS']['scaled'] = pandas.DataFrame(s).T
dataset['ELS']['scaled'].index = dataset['ELS']['default'].index

min_max_scaler = preprocessing.MinMaxScaler()
s = min_max_scaler.fit_transform(dataset['MAG']['default'])
dataset['MAG']['scaled'] = pandas.DataFrame(s)
dataset['MAG']['scaled'].index = dataset['MAG']['default'].index
```

```python
def get_hourly_ticks(indices):

    last_hour = indices[0].hour
    hourly_ticks = list()

    for i, index in enumerate(indices):
        if index.hour != last_hour:
            hourly_ticks.append([i, index])
            last_hour = index.hour
    return hourly_ticks
```

```python
image = dict()
image['IMA'] = dataset['IMA']['scaled'].T
image['ELS'] = dataset['ELS']['scaled'].T

fig, ax = plt.subplots(3,1,figsize=(20,10))
im = dict()
im[0] = ax[0].imshow(image['IMA'], interpolation='nearest', aspect='auto')
im[1] = ax[1].imshow(image['ELS'], interpolation='nearest', aspect='auto')
im[2] = ax[2].plot(dataset['MAG']['scaled'])

ticks = dict()
ticks['IMA'] = get_hourly_ticks(dataset['IMA']['scaled'].index)
ticks['ELS'] = get_hourly_ticks(dataset['ELS']['scaled'].index)
ticks['MAG'] = get_hourly_ticks(dataset['MAG']['default'].index)

ax[0].set_xticks( [i[0] for i in ticks['IMA']] )
ax[0].set_xticklabels(["{}h".format(Y[1].hour) for Y in ticks['IMA']])

ax[1].set_xticks( [i[0] for i in ticks['ELS']] )
ax[1].set_xticklabels(["{}h".format(Y[1].hour) for Y in ticks['ELS']])

ax[2].set_xticks( [i[1] for i in ticks['MAG']] )
ax[2].set_xticklabels(["{}h".format(Y[1].hour) for Y in ticks['MAG']])

ax[2].set_xlim(dataset['IMA']['scaled'].index[0], dataset['IMA']['scaled'].index[-1])
```

```python
# plotting various linear traces
data = dataset['default']
transposed = data.T
T = data.index

plt.figure(figsize=(20,10))
trace = transposed.max()
plt.plot(trace/trace.max())
plt.xlim([T[0], T[len(T)-1]])
plt.show()

plt.figure(figsize=(20,10))
trace = transposed.mean()
plt.plot(trace/trace.max())
plt.xlim([T[0], T[len(T)-1]])
plt.show()

plt.figure(figsize=(20,10))
trace = transposed.var()
plt.plot(trace/trace.max())
plt.xlim([T[0], T[len(T)-1]])
plt.show()
```
